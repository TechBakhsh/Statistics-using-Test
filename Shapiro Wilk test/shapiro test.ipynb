```python
# example of the shapiro wilk test normalization
from scipy.stats import shapiro
data = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810, -1.655]
shapiro(data)
```

```python
stat,p = shapiro(data)
print("stat =",stat)
print("p  =", p)
if p>0.05:
    print("this is normalize data")
else:
    print("this data is not normalize")
```

# 2- Correlation test
1-pearso Correlation test Cofficient
* test whether two  samples have a Linear relationship
  # Assumption 
  1- Observe each samples are independent and indentically distributed (lid)

  2-  Observe each samples are nomally distributed

  3-  Observe each samples are same varience 

  4- Interpretation

  * HO: the two samples are independent
  * HT: there is dependency between the samples 

# 1- Spearmanr  coorelation test

```python
# example of  spearmanr correlation test
from scipy.stats import spearmanr
data1 = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810,]
data2 = [0.128,-7.458, -5.780, -0.230, -3.056, -3.090, -1.340, -1.670, -1.810,]
p,stats=spearmanr(titanic["age",titanic["fare"]])
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably dependent data (coorelation exist)")
else:
    print("probably independent data(no correlation)")
    
```

# 2- Pearsonar Correlation test

```python
# example of  pearsonr correlation test
from scipy.stats import pearsonr
data1 = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810,]
data2 = [0.128,-7.458, -5.780, -0.230, -3.056, -3.090, -1.340, -1.670, -1.810,]
p,stats=pearsonr(data1, data2)
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably dependent data (coorelation exist)")
else:
    print("probably independent data(no correlation)")
```

# 3- Chi- Squared Test 
* test whether two categorial variables are independent

```python
# example of  Chi-Squared test
from scipy.stats import chi2_contingency
data1 = [[10,20,30,30],[6,9,17,0]]
stats,p, dof, expected =chi2_contingency(data1)
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably dependent ")
else:
    print("probably independent ")
```

# 4- parametric statiscal hypothesis test

1- Student t-test 
* tests whether the two means sample are independent are significant different

```python
# example of student t-test 
from scipy.stats import ttest_ind
data1 = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810,]
data2 = [0.128,-7.458, -5.780, -0.230, -3.056, -3.090, -1.340, -1.670, -1.810,]
p,stats=ttest_ind(data1, data2)
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably same data")
else:
    print("probably differnt data")
```

2- paired student test
* test  whether the two means sample are independentare significant different

```python
# example of student t-test 
from scipy.stats import ttest_rel
data1 = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810,]
data2 = [0.128,-7.458, -5.780, -0.230, -3.056, -3.090, -1.340, -1.670, -1.810,]
p,stats=ttest_rel(data1, data2)
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably same data")
else:
    print("probably differnt data")
```

# Analysis of varience test (ANOVA)
* test  whether the two means sample are independentare significant different

```python
# example of anova
from scipy.stats import f_oneway
data1 = [0.128,-0.458, -0.780, -0.230, 0.056, 0.090, -1.340, -1.670, -1.810,]
data2 = [0.128,-7.458, -5.780, -0.230, -3.056, -3.090, -1.340, -1.670, -1.810,]
data3= [0.128,-7.458, -5.780, -3.230, -5.056, -3.090, -1.340, -1.670, -1.810,]
p,stats=f_oneway(data1, data2, data3)
print("p =", p)
print("stats=", stats)
if p>0.05:
    print("probably same data")
else:
    print("probably differnt data")
```
